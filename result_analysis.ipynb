## Processed missing with MEAN

```python
import numpy as np
import pickle 

k=0
range_of_dropout = [p/10 for p in range(0, 9)]
list_of_mean_val_losses = list()
list_of_mean_val_acc = list()
list_of_mean_train_losses = list()
list_of_mean_train_acc = list()

for dropout in range_of_dropout:
    with open('Results\\processed_missing\\mean\\dict_of_k_results_cv{}_dropout-{}_3-layered_Adam_epochs_200_lr_0.001.pkl'
    .format(CV, dropout), 'rb') as handle:
        dict_of_results = pickle.load(handle)
    
    print('Dropout rate: {}'.format(dropout))
    print('Estimated average val loss: {}'.format(np.mean(dict_of_results[k]['val_loss'])))
    print('Estimated average val std loss: {}'.format(np.std(dict_of_results[k]['val_loss'])))
    print('Estimated average val acc: {}'.format(np.mean(dict_of_results[k]['val_acc'])))
    print('Estimated average val std acc: {}'.format(np.std(dict_of_results[k]['val_acc'])))
    print('Estimated average convergence time: {}'.format(np.mean(dict_of_results[k]['convergence_time'])))
    
    list_of_mean_train_acc.append(np.mean(dict_of_results[k]['train_acc']))
    list_of_mean_val_acc.append(np.mean(dict_of_results[k]['val_acc']))
    list_of_mean_train_losses.append(np.mean(dict_of_results[k]['train_loss']))
    list_of_mean_val_losses.append(np.mean(dict_of_results[k]['val_loss']))
    

list_of_mean_val_acc_processed_missing_mean = list_of_mean_val_acc
list_of_mean_val_loss_processed_missing_mean = list_of_mean_val_losses
```

```python
plt.plot(range_of_dropout, list_of_mean_train_acc, label = "Train acc")
plt.plot(range_of_dropout, list_of_mean_val_acc, label = "Val acc")

plt.xlabel('Dropout rate')
plt.ylabel('Accuracy')
plt.gca().invert_xaxis()

plt.legend()
```

```python
plt.plot(range_of_dropout, list_of_mean_train_losses, label = "Train loss")
plt.plot(range_of_dropout, list_of_mean_val_losses, label = "Val loss")

plt.xlabel('Dropout rate')
plt.ylabel('Mean Negative-Log Likelihood Loss')
plt.gca().invert_xaxis()

plt.legend()
```
